```python
import numpy as np
#import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)
X = np.array([x,y])
```

```python
plt.figure()
plt.scatter(X[0],X[1])
plt.show()
```

```python
w = np.diag([2,.5])                    # eigen Values diagonal matrix
v = np.array([[1,1], [-1,1]]).T        # eigen Vectors, should be orthogonal
V = v/np.linalg.norm(v, axis=0)        # eigen Vectors normalized
A = np.matmul(np.matmul(V, w), V.T)   # A = V * w * V.T
#A = np.matmul(V, w)                    # A = V * w , as X is already aligned to principle axis
A, V
```

```python
A1 = [[1.25,-.75],[-.75,1.25]]
#A1 = A
w1,V1 = np.linalg.eig(A1)
w1,V1
```

```python
X1 = np.matmul(A1, X)
plt.figure(figsize=(9, 8), dpi= 80, facecolor='w', edgecolor='k')
plt.scatter(X[0],X[1] )
plt.scatter(X1[0],X1[1], 20, 'r', 'x')
plt.show()
```

```python
E = np.matmul(X1,X1.T)/N
np.matmul(X,X.T)/N,np.matmul(X1,X1.T)/N
```

```python
w2,v2 = np.linalg.eig(E)
w2**.5,v2
```
